# **N-Grams in NLP: Understanding Word Sequences**

## **1Ô∏è‚É£ What are N-Grams?**
An **N-gram** is a **sequence of N words** from a given text. It helps in analyzing text structure, understanding phrase frequency, and improving **machine learning models** in NLP.

### **üîπ Types of N-Grams**
| N-gram Type  | Example (for "machine learning is powerful") |
|-------------|----------------------------------|
| **Unigram (1-word)**  | "machine", "learning", "is", "powerful" |
| **Bigram (2-words)**  | "machine learning", "learning is", "is powerful" |
| **Trigram (3-words)** | "machine learning is", "learning is powerful" |

üìå **Why Use N-Grams?** \
‚úî Identifies frequent phrases in a text  
‚úî Useful for **text analysis, autocomplete, and machine translation**  
‚úî Helps **feature engineering** for NLP models

## **2Ô∏è‚É£ Generating N-Grams in Python**
We will generate **Unigrams, Bigrams, and Trigrams** using **NLTK** and visualize them using **Matplotlib**.


______________

```python
# Import libraries :
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample text tokenized into words
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
```

```python
print(tokens)
```

## Unigram Analysis (n=1) :

```python
# Generate unigrams
unigrams = pd.Series(nltk.ngrams(tokens, 1)).value_counts()

# Display the top 10 most common unigrams
print(unigrams.head(10))

# Plot top 10 unigrams
unigrams[:10].sort_values().plot.barh(color='lightsalmon', width=0.9, figsize=(12, 6))
plt.title('10 Most Frequently Occurring Unigrams')
plt.xlabel('# of Occurrences')
plt.ylabel('Unigram')
plt.show()
```

## Bigram Analysis (n=2) :

```python
# Generate bigrams
bigrams = pd.Series(nltk.ngrams(tokens, 2)).value_counts()

# Display the top 10 most common bigrams
print(bigrams.head(10))

# Plot top 10 bigrams
bigrams[:10].sort_values().plot.barh(color='skyblue', width=0.9, figsize=(12, 6))
plt.title('10 Most Frequently Occurring Bigrams')
plt.xlabel('# of Occurrences')
plt.ylabel('Bigram')
plt.show()
```

## Trigram Analysis (n=3) :

```python
# Generate trigrams
trigrams = pd.Series(nltk.ngrams(tokens, 3)).value_counts()

# Display the top 10 most common trigrams
print(trigrams.head(10))

# Plot top 10 trigrams
trigrams[:10].sort_values().plot.barh(color='lightgreen', width=0.9, figsize=(12, 6))
plt.title('10 Most Frequently Occurring Trigrams')
plt.xlabel('# of Occurrences')
plt.ylabel('Trigram')
plt.show()
```

 ## Summary :
‚úÖ Unigrams: Show individual word frequency. 

‚úÖ Bigrams & Trigrams: Capture relationships between words. 

‚úÖ N-Grams are useful for feature extraction, machine learning, and text analysis.
